```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# <u>Moving Averages</u>

```python
# Simulated noisy signal
np.random.seed(0)
time = np.arange(0, 50)
signal = np.sin(time * 0.2) + np.random.normal(scale=0.3, size=len(time))  # noisy sine wave

df = pd.DataFrame({'time': time, 'signal': signal})

# Apply Simple Moving Average with window size 5
df['SMA_5'] = df['signal'].rolling(window=5).mean()
df['SMA_3'] = df['signal'].rolling(window=3).mean()

# Plotting
plt.figure(figsize=(10, 5))
plt.plot(df['time'], df['signal'], label='Noisy Signal', alpha=0.5)
plt.plot(df['time'], df['SMA_5'], label='Smoother Signal (SMA)', linewidth=2, color='red')
plt.plot(df['time'], df['SMA_3'], label='Smoothed Signal (SMA)', linewidth=2, color='green')
plt.xlabel('Time')
plt.ylabel('Signal')
plt.title('Denoising with Simple Moving Average')
plt.legend()
plt.grid(True)
plt.show()
```
